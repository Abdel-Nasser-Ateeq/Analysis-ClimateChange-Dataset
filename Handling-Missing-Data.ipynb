```python
def missing_ratio(df):
    DataFrame = df.isna().sum().rename('Num_0f_missing')

    # Rename the index to Feature
    DataFrame = DataFrame.reset_index().rename(columns={'index':'Feature'})

    Num_of_samples = df.shape[0]
    DataFrame['Percent_of_msising'] = DataFrame['Num_0f_missing'] / Num_of_samples

    # Select the features that have a missing values
    DataFrame = DataFrame[DataFrame['Num_0f_missing'] != 0]
    DataFrame = DataFrame.sort_values(by=['Percent_of_msising'],ascending=False)
    return DataFrame
```

```python
def handle_missing(df,list_columns,handle_type,value):
       
    if handle_type == "bfill":
        # no inplace
        df[list_columns] = df[list_columns].fillna(method='bfill')
                                    
    elif handle_type == "ffill":
        df[list_columns] = df.loc[:,list_columns].fillna(method='ffill')
        
    elif handle_type == "mean":
        df[list_columns] = df[list_columns].fillna(df.mean()).round(2)
        
    elif handle_type == "dropna0":
        df[list_columns] = df[list_columns].dropna(axis=0, how='any')
    
    elif handle_type == "dropna1":
        df[list_columns] = df[list_columns].dropna(axis=1, how='any')
        
    else: 
        df[list_columns] = df.loc[:,list_columns].fillna(value)
```
